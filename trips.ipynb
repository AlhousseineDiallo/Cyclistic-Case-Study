```python
# let's use a good function to treat the outliers

def remove_outliers(df: pd.DataFrame, column_name: str):
    """
    Removes outliers from a specified column in a DataFrame using the IQR method

    Returns:
    pd.DataFrame: A DataFrame with outliers removed from the specified column.

    """
    df_birthyear = df[df[column_name].notna()]
    Q1 = df_birthyear[column_name].quantile(0.25)
    Q3 = df_birthyear[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # let's now filter the DataFrame to remove the outliers
    filtered_df = df_birthyear[(df_birthyear[column_name] >= lower_bound) & (df_birthyear[column_name] <= upper_bound)]

    # let's reintegrate the lines without birthyear(Customer)
    final_df = pd.concat([filtered_df, df[df[column_name].isna()]])
    return final_df


```
